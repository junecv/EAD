```python
# One-Hot Encode to combine Condition1 & Condition2

      #  Artery	Adjacent to arterial street
      #  Feedr	Adjacent to feeder street	
      #  Norm	Normal	
      #  RRNn	Within 200' of North-South Railroad
      #  RRAn	Adjacent to North-South Railroad
      #  PosN	Near positive off-site feature--park, greenbelt, etc.
      #  PosA	Adjacent to postive off-site feature
      #  RRNe	Within 200' of East-West Railroad
      #  RRAe	Adjacent to East-West Railroad

lst = list(['Artery', 'Feedr', 'Norm', 'RRNn', 'RRAn', 'PosN', 'PosA', 'RRNe', 'RRAe'])
train_condition = pd.DataFrame()

for i in lst:
  a = train['Condition1'].str.contains(i) | train['Condition2'].str.contains(i)
  train_condition = pd.concat([train_condition, a], axis=1)

C_lst = ['C_' + i for i in lst]
train_condition.columns = C_lst
train_condition.head()

# combine dataframes
train = pd.concat([train, train_condition], axis=1)
```

### Model fine tuning - Poly (I) with Ridge

```python
# Since Polynomial Regression (I) - original & ridge regularization 
# gave the best CV result, below fine-tuning Ridge Regression

from sklearn.model_selection import train_test_split

drop_lst = ['Id','SalePrice', 'SalePrice_MMS', 'SalePrice_StdS', 'E_PreCast', 'RoofMatl_ClyTile']

X = train_c.drop(drop_lst,axis=1, inplace=False)                             # independant features
y = train_c['SalePrice_StdS']					                                      # dependant variable

# split kaggle trainig into training and validation and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)
```

```python
from sklearn.linear_model import Ridge

ridge_results = pd.DataFrame(columns=['i', 'ridge_rmse', 'ridge_mean'])
grid = np.linspace(0,100,11)

for i in grid:
  ridge = Ridge(alpha=i, solver='cholesky')
  ridge.fit(X_train,y_train)
  pred = ridge.predict(X_train)
  ridge_mse = mean_squared_error(y_train, pred)
  ridge_rmse = np.sqrt(ridge_mse)
  score = cross_val_score(ridge, X_train, y_train, scoring='neg_mean_squared_error', cv=10)
  ridge_CV_score = np.sqrt(-score)
  ridge_mean = ridge_CV_score.mean()
  result = pd.Series(list([i, ridge_rmse, ridge_mean]), index=ridge_results.columns)
  ridge_results = ridge_results.append(result, ignore_index=True)
```

## Conclusion

    The prediction achieved score of 0.13051 in Kaggle leader board. The key assumption here is that the response distribution of test set is the same as the training set. This may not be the case in real life.

    After review base model and two polynomial regression model, it is found that fine-tuning parameter gives significant improvement on the results. Also, cleaning the dataset, filling in missing data, and removing outliers helped develop a better model in this case. 

    Polynomial regression with n=2 is overfitting training data, while selective compounding of highly correlated features can avoid this effect. 

    **Learning**

    Piece-wise model is the way to further improve the model. Base on residual plots, linear model (polynomial or not) tend to miss the mark on "edges" where response is further away from the mean (beyond -1 / 1 standard deviation). My next attempt will be learning technically to create piece-wise model.
    
    Interestingly, a better performaing Ridge model (i = 12.6 vs i = 13) based on CV result did not produce better performance in the Kaggle Leader board.
